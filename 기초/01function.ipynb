#### 파이썬 함수의 구조


```
def 함수의이름(매개변수)
  # 들여쓰기를 계속 유지할 것
  <수행할 문장 1>
  <수행할 문장 2>
  return a + b
```

- 매개변수라는 '입력값'을 가진다
- return값 이라는 '출력값'을 가진다
- 입력값과 출력값이 필수는 아니다! (없어도 된다)

````
def SayHelloFunction():
  print("안녕하세요!")

for i in range(10):
  SayHelloFunction()
````

````
def SayHelloFunction2(name):
  print("안녕하세요! " + name + "님")

for name in ["벼리", "철수", "미영", "영미", "동글"]:
  SayHelloFunction2(name)
````
